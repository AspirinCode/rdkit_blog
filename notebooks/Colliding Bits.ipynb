# Looking at bit collisions in Morgan fingerprints

In the group over the past few days we've had a few conversations about bit collisions in Morgan fingerprints.

This post is an exploration of how often that happens.

```python
from rdkit import Chem,DataStructs
import time,random
import numpy as np
from collections import defaultdict
from rdkit.Chem import Draw
from rdkit.Chem import rdMolDescriptors as rdmd
from rdkit.Chem.Draw import IPythonConsole
from rdkit import DataStructs
from rdkit import rdBase
from __future__ import print_function

print(rdBase.rdkitVersion)
import time
print(time.asctime())
```

For test data I'll use the ChEMBL compounds from 2010-2012 that I built in an earlier post.

```python
filen='../data/chembl16_2010-2012.smi'
!wc -l $filen
!head $filen
```

Loop over the molecules and build fingerprints of multiple radii and hashing lengths.

```python
counts=defaultdict(list)
for i,line in enumerate(file(filen)):
    m = Chem.MolFromSmiles(line.split()[0])
    if not m: continue
    for v in 0,1,2,3:
        counts[(v,-1)].append(len(rdmd.GetMorganFingerprint(m,v).GetNonzeroElements()))
        for l in 1024,2048,4096,9192:
            counts[(v,l)].append(rdmd.GetMorganFingerprintAsBitVect(m,v,l).GetNumOnBits())
    if not (i+1)%5000:
        print("Done {0}".format(i+1))
```

Now plot histograms of the numbers of collisions along with the fractions of collisions

```python
figure(figsize=(16,20))

pidx=1
#----------------------------
for nbits in (9192,4096,2048,1024):
    subplot(4,2,pidx)
    pidx+=1
    v1=np.array(counts[1,-1])
    v2=np.array(counts[1,nbits])
    d1 = v1-v2
    d1p=np.array(d1,np.float)
    d1p/=v1
    v1=np.array(counts[2,-1])
    v2=np.array(counts[2,nbits])
    d2 = v1-v2
    d2p=np.array(d2,np.float)
    d2p/=v1
    v1=np.array(counts[3,-1])
    v2=np.array(counts[3,nbits])
    d3 = v1-v2
    d3p=np.array(d3,np.float)
    d3p/=v1
    
    _=hist((d1,d2,d3),bins=20,log=True,label=("r=1","r=2","r=3"))
    title('%d bits'%nbits)
    _=legend()
    
    subplot(4,2,pidx)
    pidx+=1
    _=hist((d1p,d2p,d3p),bins=20,log=True)
    _=hist((d1p,d2p,d3p),bins=20,histtype='step',cumulative=-1,normed=True)


```

So, there are definitely some collisions.

```python
random.seed(0xF00D)
smis = [x.split()[0] for x in file(filen)]
ivs=[random.randint(0,len(smis)-1) for x in range(10000)]
jvs=[random.randint(0,len(smis)-1) for x in range(10000)]
pairs=zip(ivs,jvs)

sims=defaultdict(list)
for i,j in pairs:
    mi = Chem.MolFromSmiles(smis[i])
    if not mi:
        continue
    mj = Chem.MolFromSmiles(smis[j])
    if not mj:
        continue
    for v in 0,1,2,3:
        fpi=rdmd.GetMorganFingerprint(mi,v)
        for k,v in fpi.GetNonzeroElements().items():
            fpi[k]=1
        fpj=rdmd.GetMorganFingerprint(mj,v)
        for k,v in fpj.GetNonzeroElements().items():
            fpj[k]=1
        sims[(v,-1)].append(DataStructs.TanimotoSimilarity(fpi,fpj))
        for l in 1024,2048,4096,9192:
            fpi=rdmd.GetMorganFingerprintAsBitVect(mi,v,l)
            fpj=rdmd.GetMorganFingerprintAsBitVect(mj,v,l)
            sims[(v,l)].append(DataStructs.TanimotoSimilarity(fpi,fpj))
```

```python
figure(figsize=(12,20))

pidx=1
#----------------------------
for nbits in (9192,4096,2048,1024):
    subplot(4,1,pidx)
    pidx+=1
    v1=np.array(sims[1,-1])
    v2=np.array(sims[1,nbits])
    d1 = v1-v2
    d1p=np.array(d1,np.float)
    d1p/=v1
    v1=np.array(sims[2,-1])
    v2=np.array(sims[2,nbits])
    d2 = v1-v2
    d2p=np.array(d2,np.float)
    d2p/=v1
    v1=np.array(sims[3,-1])
    v2=np.array(sims[3,nbits])
    d3 = v1-v2
    d3p=np.array(d3,np.float)
    d3p/=v1
    
    _=hist((d1,d2,d3),bins=20,log=True,label=("r=1","r=2","r=3"))
    title('%d bits'%nbits)
    _=legend()
    

```
